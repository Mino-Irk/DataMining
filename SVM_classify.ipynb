# tf-idfを用いたテキスト分類お試し版

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import numpy as np

categories = ['alt.atheism', 'soc.religion.christian','comp.graphics', 'sci.med', 'rec.sport.baseball']

all_data= fetch_20newsgroups(shuffle=True, categories=categories, random_state=42)


# tfidfによる重み付け
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(all_data.data)

tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

X_train_tfidf.toarray()

all_X = X_train_tfidf
all_y = all_data.target
print(all_X)
print(len(all_data.target))
```

## 不要な活用語をカットする

```python
# データの中身をlistで入力

file_name = './morpheme_list'
file_list = os.listdir(file_name)
all_datas = []
all_ys = []
count = 0

# 分類ディレクトリを取得
for file in file_list:
    # ディレクトリかどうか判別
    if os.path.isfile(file_name+'/'+file) == True:
        continue
        
    files = os.listdir(str(file_name + '/' + file))
    
    print(file)
    # 分類ごとのnovelデータ取得
    for fil in files:
        print(fil)
        
        fi = str(file_name + '/' + file + '/' + fil)                                
        id = open(fi, 'r') 
        conte = ''
                                    
        ele = fil.split('.')                                                   
        #print(ele[0])  #本のid                                                      
        for contents in id:                                                         
            ele2 = contents.split(',')                                              
            if(ele2[1]!="助詞"):                                                    
                if(ele2[1]!="副詞"):                                                
                    if(ele2[1]!="助動詞"):                                          
                        if(ele2[1]!="記号"):                                        
                            if(ele2[1]!="接続詞"):                                  
                                if(ele2[0]!='——'): 
                                    conte = str(conte + ele2[0] + " ")                                                                         
        # 1nobelごとにlistへ追加
        #print(conte)
        all_datas.append(conte)
        all_ys.append(count)
        
    count = count + 1
```

```python
#list型からndarray型へ変換
#all_data = np.array([str(a.readline()),str(b.readline()),str(c.readline()),str(d.readline()),str(e.readline()),str(f.readlines()),str(g.readlines()),str(h.readline()),str(i.readline())])
all_data = np.array(all_datas)

# tfidfによる重み付け
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(all_data)

tfidf_transformer = TfidfTransformer(use_idf=True, norm='l2',smooth_idf=True)
#np.set_printoptions(precsion=2)

X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.toarray()

all_X = X_train_tfidf
print(all_X)
# list型をndarray型へ変換
#all_y = np.array([0,1,2,3,4,5,6,7,9])
all_y = np.array(all_ys)
print(all_y)
```

# SVM分類

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics import f1_score
from sklearn import svm
from sklearn.svm import SVC

train_X, test_X, train_y, test_y = train_test_split(all_X, all_y,
                                                    test_size=0.2,
                                                    random_state=42)


clf2 = svm.SVC(kernel='rbf', C=5, gamma=0.2)
clf2.fit(train_X,train_y)
pred_y2 = clf2.predict(test_X)
print(f1_score(test_y, pred_y2, average='macro'))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(test_y, pred_y2))
```
